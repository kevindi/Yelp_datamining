```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as sk_data
import sklearn.metrics as metrics

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline
import json
```

```python
datas = []

with open('data/yelp_academic_dataset_business.json') as b_json:
    for each_class in b_json:
        data = json.loads(each_class)
        row = []
        if 'Phoenix' in data['city'] and 'Restaurants' in data['categories']:
            row.append(data['business_id'])
            row.append(data['categories'])
            row.append(data['latitude'])
            row.append(data['longitude'])
            row.append(data['review_count'])
            row.append(data['stars'])
            row.append(data['open'])
            
            if 'Accepts Credit Cards' in data['attributes']:
                row.append(data['attributes']['Accepts Credit Cards'])
            else:
                row.append('N/A')
            
            if 'Alcohol' in data['attributes']:
                row.append(data['attributes']['Alcohol'])
            else:
                row.append('N/A')
            
            if 'Attire' in data['attributes']:
                row.append(data['attributes']['Attire'])
            else:
                row.append('N/A')
            
            if 'Delivery' in data['attributes']:
                row.append(data['attributes']['Delivery'])
            else:
                row.append('N/A')
            
            if 'Good For Groups' in data['attributes']:
                row.append(data['attributes']['Good For Groups'])
            else:
                row.append('N/A')
            
            if 'Good for Kids' in data['attributes']:
                row.append(data['attributes']['Good for Kids'])
            else:
                row.append('N/A')
                
            if 'Has TV' in data['attributes']:
                row.append(data['attributes']['Has TV'])
            else:
                row.append('N/A')
                
            if 'Noise Level' in data['attributes']:
                row.append(data['attributes']['Noise Level'])
            else:
                row.append('N/A')
                
            if 'Outdoor Seating' in data['attributes']:
                row.append(data['attributes']['Outdoor Seating'])
            else:
                row.append('N/A')

            if 'Price Range' in data['attributes']:
                row.append(data['attributes']['Price Range'])
            else:
                row.append('N/A')
                
            if 'Takes Reservations' in data['attributes']:
                row.append(data['attributes']['Takes Reservations'])
            else:
                row.append('N/A')
                
            if 'Waiter Service' in data['attributes']:
                row.append(data['attributes']['Waiter Service'])
            else:
                row.append('N/A')
                
            if 'Wheelchair Accessible' in data['attributes']:
                row.append(data['attributes']['Wheelchair Accessible'])
            else:
                row.append('N/A')
        
            if 'Wi-Fi' in data['attributes']:
                row.append(data['attributes']['Wi-Fi'])
            else:
                row.append('N/A')
                
            if 'Take-out' in data['attributes']:
                row.append(data['attributes']['Take-out'])
            else:
                row.append('N/A')
            
            datas.append(row)
```

```python
business_df_columns = ['business_id', 'categories', 'latitude', 'longitude', 'review_count', 'stars', 'open', 'Accepts_Credit_Cards'
                      , 'Alcohol', 'Attire', 'Delivery', 'Good_For_Groups', 'Good_for_Kids', 'Has_TV', 'Noise_Level'
                      , 'Outdoor_Seating', 'Price_Range', 'Takes_Reservations', 'Waiter_Service', 'Wheelchair_Accessible'
                      , 'Wi_Fi', 'Take_out']
```

```python
len(business_df_columns)
```

```python
business_df = pd.DataFrame(data=datas, columns=business_df_columns)
```

```python
review_datas = []
review_datas_all = []

with open('data/yelp_academic_dataset_review.json') as b_json:
    for each_class in b_json:
        data = json.loads(each_class)
        review_datas.append([data['business_id'], data['text'], data['stars']])
        review_datas_all.append(data)
        break;
```

```python
review_datas_all
```

```python
review_datas_df = pd.DataFrame(review_datas, columns=['business_id', 'review_text', 'stars'])
```

```python
review_datas_df
```

```python
review_data_df = pd.DataFrame(review_datas_df.groupby('business_id')['review_text'].apply(list))
```

```python
review_data_df['mean_star'] = review_datas_df.groupby('business_id')['stars'].mean()
```
